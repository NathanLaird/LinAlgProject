```python
def euclidean_dist(X,Y):
    if X.shape != Y.shape:
        raise Exception('These Vectors Should Have The Same Size! Please Try Again.')
    if X.shape[1] != 1:
        raise Exception('These Vectors Should Be Column Vectors! Please Try Again.')
    return np.linalg.norm(X-Y)
    
```

```python
def cosine_sim(X,Y):
    return (np.sum((np.inner(X,Y)))/(np.linalg.norm(X)*np.linalg.norm(Y)))
    
```

```python
XX = np.array([[0],[1]])
ZZ = np.array([[1],[0]])

cosine_sim(XX,ZZ)
```

```python
def compute_dist(data_matrix,Function):
    mean_arr = np.mean(data_matrix, axis = 0)
    mean_arr = np.array([mean_arr])
    dist_arr = []
    
    for vec in data_matrix:
        vec = np.array([vec])
        dist_arr.append(Function(vec.T,mean_arr.T))
    return (dist_arr)
    
```

```python
euclidean_dists = compute_dist(iris_data, euclidean_dist)
```

```python
plt.hist(euclidean_dists)
```

```python
cosine_sims = compute_dist(iris_data, cosine_sim)
```
