```python
import numpy as np
import matplotlib.pyplot as plt
#Hi Jack
%matplotlib inline
```

```python
StochMatrix = np.array([[0.7,0.1,0.0],[0.2,0.9,0.2],[0.1,0,0.8]])
```

```python
TwoThousandFour = np.array([[0.25],[0.20],[0.55]])
```

```python
TwoThousandNine = StochMatrix @ TwoThousandFour
print(TwoThousandNine)
```

```python
TwoThousandFourteen = StochMatrix @ TwoThousandNine
print(TwoThousandFourteen)
```

```python
#Residential
num = 20
C = []
I = []
R = []
Vec = np.array([[0],[0],[1]])
for i in range(num):
    C.append(Vec[0,0])
    I.append(Vec[1,0])
    R.append(Vec[2,0])
    Vec = StochMatrix @ Vec

plt.plot(C)
plt.plot(I)
plt.plot(R)


```
